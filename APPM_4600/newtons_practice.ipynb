```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (x**2)

def fp(x):
    return(2*x)

def newtons(f,fp,x0,Nmax,tol):
    N = 0
    if fp(x0) == 0:
        print("Error: f'(x)=0. Method Failed. Try another x0 or method.")
        print("Method failed when n = ", N, ", x = ", x0)
    x1 = x0 - f(x0)/fp(x0)
    N = 1
    seq = []

    while np.abs(x1-x0)>tol and N<Nmax:
        seq.append(x0)
        x0 = x1
        x1 = x1 = x0 - f(x0)/fp(x0)
        N = N+1

        if fp(x1) == 0:
            print("Error: f'(x)=0. Method Failed. Try another x0 or method.")
            print("Method failed when n = ", N,", x = ", x1)
            break
    return x1,N,seq

x0 = 1
Nmax = 100
tol = 1e-9

x1,N,seq = newtons(f,fp,x0,Nmax,tol)
print("x = ", x1, ", using n = ",N," iterations")

plt.figure()
plt.plot(seq)
plt.yscale("log")
plt.xlabel("Iterations")
plt.ylabel("Error")
plt.grid("minor")

def f(x):
    return (-x**3+4*x+5)

def fp(x):
    return(-3*x**2+4)

x0 = 2
x1,N,seq2 = newtons(f,fp,x0,Nmax,tol)
print("x = ", x1, ", using n = ",N," iterations")

plt.figure()
plt.plot(seq2)

plt.xlabel("Iterations")
plt.ylabel("Error")
plt.grid("minor")

```
