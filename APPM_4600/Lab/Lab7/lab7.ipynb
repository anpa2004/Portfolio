```python
import numpy as np
import matplotlib.pyplot as plt
import random as ran
import mypkg.my2DPlotB
import math
from numpy.linalg import inv 
    
def vandermonde(x):
    n = len(x)
    v = np.empty((n,n))
    # i = row, j = col
    for i in range(n):
        for j in range(n):
            v[i,j] = x[i]**j
    return v

def interp_coefficients(v,y):
    a = np.linalg.solve(v,y)
    return a

def solve_coeff(a,xq):
    p = []
    n = len(a)
    m = len(xq)
    for i in range(m):
        pn = 0
        for j in range(n):
            pn = a[j]*xq[i]**j + pn
        p.append(pn)
    return(p)

f = lambda x: solve_coeff(a,x)
        
n = 15
x = [100*ran.random() for _ in range(n)]
y = [100*ran.random() for _ in range(n)]
xq = np.array([1,np.pi,45,0.2])

xempty = np.linspace(0,100,1000)
#x = np.array([1,3,5,14])
#y = np.array([0,2,-1,-6])

v = vandermonde(x)
a = interp_coefficients(v,y)
p = solve_coeff(a,xq)
print(p)

plt.figure()
plt.scatter(x,y)
plt.plot(xempty,f(xempty))

    
```
